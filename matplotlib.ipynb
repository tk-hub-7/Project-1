```python
from matplotlib import pyplot as plt
```

```python
#plotting
x=[1,2,3]
y=[1,2,3]
plt.plot(x,y)
```

```python
#naming the plot and labeling the axes
x=[1,2,3]
y=[4,3,5]
plt.plot(x,y)
plt.title("My plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid()
```

```python
#adding line color and width
x=[1,2,3]
y=[5,2,1]
plt.plot(x,y,color='r',linewidth=3)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid()
```

```python
#multiplot
x=[1,2,3]
y=[5,4,1]

x1=[4,3,2]
y1=[2,3,1]

plt.plot(x,y,color='r',label="line1",linewidth="2")
plt.plot(x1,y1,color='g',label="line2",linewidth="3")

plt.title("Multiplot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
```

```python
#multiple plots example
mon=[780,345,567,789,345,236,789]
tue=[467,678,996,356,235,676,356]
wed=[245,785,795,800,239,789,987]

time_hrs=[6,7,8,9,10,11,12]

plt.plot(time_hrs,mon,color='r',linewidth="2",linestyle="dotted",label="monday")
plt.plot(time_hrs,tue,color="g",linewidth="3",linestyle="dotted",label="tuesday")
plt.plot(time_hrs,wed,color='k',linewidth="4",linestyle="dotted",label="wednesday")

plt.title("website users")
plt.xlabel("time spent")
plt.ylabel("days")
plt.legend()
plt.grid()
```

```python
#sub plot--

x=[1,2,3,4,5]
y=[4,3,2,1,8]
z=[5,3,1,9,7]

fig,axs=plt.subplots(2,2)

axs[0][0].plot(x,y,color='r')
axs[0][1].plot(z,y,color='b')
axs[1][0].plot(x,z,color='k')
axs[1][1].plot(y,z,color='g')

axs[0][0].set(xlabel='x-axis',ylabel='y-axis',title="first")
axs[0][1].set(xlabel='x-axis',ylabel='y-axis',title="second")

axs[1][0].set(xlabel='x-axis',ylabel='y-axis',title="third")
axs[1][1].set(xlabel='x-axis',ylabel='y-axis',title="fourth")
```

```python
#Alpha property--- used to adjust the transparency of the graph plot using alpha attribute
x=[1,2,3]
y=[5,6,1]

plt.plot(x,y,linewidth="4",alpha=0.1) #default value of alpha is 1 
```

```python
#Bar graph--- used to compare categorical data

x=['mercury','venus','earth','mars']
y=[2.9,6.7,9.8,9.5]

plt.bar(x,y,color='g')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Acceleration due to gravity")
```

```python
#Histogram--- used to represent statistical information that uses rectangles to show the frequency of 
#dataitems in a successive numerical intervals of size.

#example--population ages watching a movie

population_ages=[0,10,20,23,24,25,26,27,28,29,30,35,40,42,44,45,48,50,55,57,60,65,70]

#plt.hist(population_ages,histtype="bar",rwidth=0.8)
plt.hist(population_ages,histtype="bar",rwidth=0.4,bins=[0,20,40,60,80])
plt.title("Ages of people watching a movie")
plt.xlabel("ages")
plt.ylabel("weight")
```

```python
#Area plot---represents change in quantities over time
#Example- time spent by a student

days=[1,2,3,4,5]
sleeping=[7,8,9,8,7]
playing=[3,4,5,3,2]
school=[6,5,6,7,6]

plt.plot([],[],color='r',label="sleeping")
plt.plot([],[],color='c',label="playing")
plt.plot([],[],color='b',label="school")

plt.stackplot(days,sleeping,playing,school,colors=['r','c','b'])
plt.title("Time spent by student")
plt.xlabel("days")
plt.ylabel("time")
plt.legend()
```

```python
#Pie chart-- used to display data in a circular graph.
hours=[6,8,3,4]
activities=["school","sleeping","eating","playing"]
cols=['r','c','y','g']

plt.pie(hours,labels=activities,colors=cols,shadow=True,explode=[0.1,0,0,0],autopct="%.2f")
plt.title("PIE CHART")
```

```python
#Scatter plot-- used to represent the relationship between two sets of data

height=[4,4.5,5,5.2,5.3,5.6,6]
weight=[40,50,54,55,57,62,65]

plt.scatter(height,weight,label="scatter",s=50,color='y',marker="*")#s represents the marker size
plt.title("Scatter plot")
plt.xlabel("height")
plt.ylabel("weight")
plt.legend()
```
