```python
import numpy as np
import pandas as pd

data={'Team':['Team1','Team2','Team3','Team2','Team1','Team2','Team2','Team2'],'Name':['karthik','kapil','akshith','nivas','srawik','mani','shivasai','ganesh'],'Age':[19,20,19,20,21,21,20,19],'Address':['hanamkonda','hyderabad','warangal',
                   'hanamkonda','warangal','hyderabad','siddipet','khammam'],'Year':[2020,2019,2019,2020,2018,2020,2019,2018],}
df=pd.DataFrame(data)
df
```

```python
gb=df.groupby('Address')
gb
```

```python
gb.get_group('hanamkonda')
```

```python
#applying group by values on two columns
gb2=df.groupby(['Team','Address'])
gb2
```

```python
gb2.get_group(('Team2','hyderabad'))
```

```python
for name,group in gb:
    display(name)
    display(group)
```

```python
gb.aggregate(np.sum) #Aggregation is a process in which we compute a summary statistic about each group.
```

```python
gb2.aggregate(np.sum) #selecting multiple columns
```

```python
#CROSS TAB-----it builds a cross-tabulation method that can,show the frequency with each certain groups of data appear.

d={'Name':['karthik','kapil','shiva','ganesh','mani','charan','yakub',
          'karthik','kapil','shiva','ganesh','mani','charan','yakub'],
   'Semester':['semester1','semester1','semester1','semester1','semester1','semester1','semester1',
               'semester2','semester2','semester2','semester2','semester2','semester2','semester2'],
   'subject':['maths','maths','maths','maths','maths','maths','maths',
              'science','science','science','science','science','science','science'],
   'result':['pass','fail','pass','pass','pass','fail','pass','pass','pass','fail','fail','pass','pass','fail']}
df=pd.DataFrame(d,columns=['Name','Semester','subject','result'])
df
```

```python
pd.crosstab(df.subject,df.result,margins=False)  #subject wise
```

```python
pd.crosstab(df.subject,df.result,margins=True)
```

```python
pd.crosstab(df.Name,df.result,margins=True)
```

```python
pd.crosstab([df.Name,df.subject],df.result,margins=True)
```
