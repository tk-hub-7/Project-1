```python
from scipy.stats import skew
import numpy as np
x=np.array([-145,-117,-119,-121,1,1,1,1,1])
skew(x)
```

```python
#Variance
import statistics
sample=[3,4,2,5,6,7,3,4,8]
print("population variance is%s"%(statistics.pvariance(sample)))
print("sample variance is %s"%(statistics.variance(sample)))
```

```python
#standard deviation
sample1=[9,2,5,4,12,7,8,11,9,3,7,4,12,5,4,10,9,6,9,4]
print("population std:",np.std(sample1))
print("sample std:",statistics.stdev(sample1))
```

```python
#co variance
x=[1,2,3,4,5]
y=[50,40,30,20,10]
cov_mat=np.stack((x,y),axis=0)
print(np.cov(cov_mat))
```

```python
# correlation
import pandas as pd
df=pd.DataFrame(list(zip(x,y)),index=['a','b','c','d','e'],columns=['x','y'])
df
```
