```python
from scipy import stats
import numpy as np
from matplotlib import pyplot as plt
import warnings
warnings.filterwarnings('ignore')
x=np.array([2,3,4,5,4,4,4,4,4])
print("mean:",np.mean(x))
print("median:",np.median(x))
print("mode:",stats.mode(x))
```
