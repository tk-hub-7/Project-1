```python
#Seaborn-- it is a datavisualization library in python which is built on top of the matplotlib
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt


#countplot--used to show count of each observation as per the category by the graph
data={'cust_id':[11,12,13,14,15,16,17,18,19],'purchased':[1,1,1,0,0,1,0,1,0],'gender':['m','f','m','f','m','m','m','m','m']}
df=pd.DataFrame(data)

#sns.countplot(x="purchased",data=df)
sns.countplot(x="purchased",data=df,hue='gender')
plt.title(" count plot")
plt.ylabel("freguency")
```

```python
sns.lineplot(x=df["station_id"],y=df["total_cost"])
```

```python
sns.scatterplot(x=df["cost_per_kwh"],y=df["station_id"])
```

```python
#rel plot-- here we can choose the lineplot or scatter plot
sns.relplot(x=df["station_id"],y=df["kwh_charged"],kind='scatter')
```

```python
#dist plot--depicts the variation in data distribution
import numpy as np
print(max(df["kwh_charged"]))
print(min(df["kwh_charged"]))

sns.distplot(df['kwh_charged'])
#sns.distplot(df['kwh_charged'],hist=False)
```

```python
#Hist plot--shows frequencys of several numerical variables
sns.histplot(df["kwh_charged"])
```

```python
df.drop(columns=['customer_id'],inplace=True)
df.head()
```

```python
df.drop(columns=['session_start_time'],inplace=True)
df.head()
```

```python
#Heat map-- uses colors to visualize the value of the matrix
corr=df.corr()
sns.heatmap(df.corr())
```

```python
#Joint plot--creates a scatter plot with two marginal histograms
print("maximum_kwh:\n",max(df['kwh_charged']))
print("minimum_kwh:\n",min(df['kwh_charged']))
print("maximum_cost:\n",max(df['cost_per_kwh']))
print("maximum_cost:\n",min(df['cost_per_kwh']))
print("mean_kwh:\n",np.mean(df['kwh_charged']))
print("mean_cost:\n",np.mean(df['kwh_charged']))

sns.jointplot(x='kwh_charged',y='cost_per_kwh',data=df)
```

```python
sns.pairplot(df)# pair plot is used to plot multiple bi variate distributions in a dataset
```
