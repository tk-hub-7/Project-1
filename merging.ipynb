```python
#Merging the dataframes
```

```python
import pandas as pd

data1={'Name':['karthik','kalyan','kapil'],'Age':[21,22,23],'Address':['hanamkonda','warangal','hyderabad']}
data2={'Name':['mani','yakub','charan','karthik'],'salary':[100000,90000,80000,70000]}
df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)

display(df1)
display(df2)
```

```python
result_inner=pd.merge(df1,df2,how='inner',on='Name')
print(result_inner)
```

```python
result_left=pd.merge(df1,df2,how='left',on='Name')
print(result_left)
```

```python
result_right=pd.merge(df1,df2,how='right',on='Name')
print(result_right)
```

```python
result_outer=pd.merge(df1,df2,how='outer',on='Name')
print(result_outer)
```

```python
#Concatenating the dataframes
```

```python
d1={'Name':['charan','nishad','izhaan','kiran'],'Age':[19,20,21,20],'Degree':['MBA','BBA','B-TECH','M-TECH']}
d2={'Name':['mani','karthik','ganesh','shivasai'],'Age':[19,20,21,20],'Degree':['B-TECH','BBA','MBA','M-TECH']}
d3={'Name':['abhi','srihari','bharath','adharsh'],'Age':[19,20,21,20],'Degree':['BBA','BBA','MBA','M-TECH']}

df1=pd.DataFrame(d1,index=[0,1,2,3])
df2=pd.DataFrame(d2,index=[4,5,6,7])
df3=pd.DataFrame(d3,index=[8,9,10,11])

display(df1)
display(df2)
display(df3)
```

```python
res=pd.concat([df1,df2,df3])
print(res)
```

```python
#RESHAPING----
PIVOT TABLE---
 A pivot table is a statistical tool that summarizes,reorganizes the rows and columns of data in a spread sheet for a desired report.
```

```python
data={'Team':['Team1','Team1','Team2','Team2','Team3','Team4'],'Year':[2019,2020,2019,2020,2021,2019],'Score':[20,30,30,30,20,20]}
df=pd.DataFrame(data)
display(df)
```

```python
table=pd.pivot_table(df,index=['Team'],values='Score',aggfunc='sum')
print(table)
```

```python
PIVOT() FUNCTION----
      The pivot() function is used to reshape the given dataframe organised by given index/column values.this function does not support aggregation.
```

```python
data={'Team':['Team1','Team1','Team2','Team2','Team3','Team4'],'Year':[2019,2020,2019,2020,2021,2019],'Score':[20,30,30,30,20,20]}
df=pd.DataFrame(data)
display(df)
piv=pd.pivot(df,index=['Team'],columns=['Year'])
display(piv)
```

```python
#hierarchial indexing----
data={'Team':['Team1','Team1','Team2','Team2','Team3','Team4'],'Year':[2019,2020,2019,2020,2021,2019],'Score':[20,30,30,30,20,20]}
df=pd.DataFrame(data)
display(df)
df.set_index(['Team','Year'],inplace=True)
display(df)
```
